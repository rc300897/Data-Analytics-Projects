```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_data= pd.read_csv("census_2011_ddw_0000b_15_2011.csv")
```

```python
census_data
```

```python
Total_census_data = census_data[(census_data["Total/Rural/Urban"]=="Total")]
Total_census_data
```

```python
Total_census_data = Total_census_data.iloc[8: , : ]
Total_census_data
```

```python
Total_census_data.columns
```

```python
Total_census_data = Total_census_data.drop(['Table Name','State Code','District Code'], axis=1)
```

```python
Total_census_data
```

```python
Total_census_data[["S","State"]]=Total_census_data["Area Name"].str.split("-",expand=True)
```

```python
Total_census_data
```

```python
Total_census_data=Total_census_data.drop(['Area Name','S'], axis =1)
```

```python
Total_census_data
```

```python
state_column= Total_census_data.pop('State')
state_column
```

```python
Total_census_data.insert(0,'State',state_column)
Total_census_data
```

```python
Total_census_Data_5 = Total_census_data.pivot(index='State',columns ='Educational Level',values=['Non-Workers - Persons','Non-Workers - Males','Non-Workers - Females','Main Activity Beggars, Vagrants etc. - Persons','Main Activity Beggars, Vagrants etc. - Males','Main Activity Beggars, Vagrants etc. - Females','Main Activity Students - Persons', 'Main Activity Students - Males',
       'Main Activity Students - Females',
       'Main Activity Household Duties - Persons',
       'Main Activity Household Duties - Males',
       'Main Activity Household Duties - Females',
       'Main Activity Dependents - Persons',
       'Main Activity Dependents - Males',
       'Main Activity Dependents - Females',
       'Main Activity Pensioners - Persons',
       'Main Activity Pensioners - Males',
       'Main Activity Pensioners - Females',
       'Main Activity Rentiers - Persons', 'Main Activity Rentiers - Males',
       'Main Activity Rentiers - Females','Main Activity Others - Persons', 'Main Activity Others - Males',
       'Main Activity Others - Females'
])
```

```python
Total_census_Data_5.columns=list(map("".join,Total_census_Data_5.columns))
Total_census_Data_5.head()
```

```python
merge_ed_pf
```

```python
merge_cen_suc = pd.merge(Total_census_data_6,merge_ed_pf, on="State")
merge_cen_suc
```

```python
poverty_data = pd.read_csv("Poverty_data.csv")
poverty_data
```

```python
poverty_data['State'].equals(merge_cen_suc['State'])
```

```python
poverty_data['State']=merge_cen_suc['State'].values
```

```python
final_merge_data=pd.merge(merge_cen_suc,poverty_data,on='State')
final_merge_data
```

```python
final_merge_data.to_csv("C:/Users/rahul/Desktop/final_merged_data.csv")
```

```python
final_merge_data.plot.scatter( x = 'Non-Workers - PersonsIlliterate', y ='No.ofPersons(lakhs) (Total)', c='red', title = "Scatter plot showing ralationship between Non of Illiterate non Workers and No of people in 100 thousands who are below Poverty Line for all states", xlabel="Non Workers Persons (Illiterate)", ylabel="No of Persons in thousands Below Poverty Line")
plt.ticklabel_format(style="plain")
plt.xticks(fontsize=8)
plt.show()
```

```python
correlation = final_merge_data['No.ofPersons(lakhs) (Total)'].corr(final_merge_data['Non-Workers - PersonsIlliterate'])
print("Correlation between illiterate Non workers and The No of people in Lakhs who are below Poverty line is " + str(correlation))
```
