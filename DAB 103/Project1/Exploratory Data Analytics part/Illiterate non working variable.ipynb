```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
census_data= pd.read_csv("census_2011_ddw_0000b_15_2011.csv")
```

```python
census_data
```

```python
Total_census_data_illiterate = census_data[(census_data["Total/Rural/Urban"]=="Total") & (census_data["Educational Level"]=="Illiterate")]
Total_census_data_illiterate
```

```python
Total_census_data_illiterate = Total_census_data_illiterate.iloc[1: , : ]
Total_census_data_illiterate
```

```python
plt.hist(Total_census_data_illiterate["Non-Workers - Persons"], bins= 10, color ='skyblue')
plt.ticklabel_format(style='plain')
plt.tick_params(axis='x', labelsize=8)
plt.xlabel("Total No of Illiterate and non working Persons")
plt.ylabel("Counts")
plt.title("Histogram showing the distribution of Illiterate and non working persons of all states")
plt.show()
```

```python
Total_census_data_illiterate["Non-Workers - Persons"].quantile([0.1,0.5,0.75,1])
```

```python
sns.kdeplot(data = Total_census_data_illiterate["Non-Workers - Persons"])
plt.ticklabel_format(style='plain')
plt.xlabel("Total No of Illiterate and non working  Persons")
plt.ylabel("Probability density of the persons")
plt.title(" Kernal density estimate plot for the No of Illiterate and non working  persons of all states ")
plt.show()
```
