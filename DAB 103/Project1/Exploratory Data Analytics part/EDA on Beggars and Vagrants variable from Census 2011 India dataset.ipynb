```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Reading the data from the csv file**

```python
census_data= pd.read_csv("census_2011_ddw_0000b_15_2011.csv")
```

### **Creating a subset of data for total count of persons for all states with total education levels**

```python
Total_census_data_State_wise = census_data[(census_data["Total/Rural/Urban"]=="Total") & (census_data["Area Name"]!="India") & (census_data["Educational Level"]=="Total")]
Total_census_data_State_wise
```

### **Plotting the bar grapgh for number of Beggars and Vagrants count in different Indian states and UTs till the year of 2011**

```python
Total_census_data_State_wise.plot(x = 'Area Name', y= 'Main Activity Beggars, Vagrants etc. - Persons', kind = 'bar')
plt.ylabel("Beggars and Vagrants count")
plt.title("Number of Beggars and Vagrants count in different states and UTs in India till the year of 2011")
plt.show()
```

### **Plotting the histogram for total number of Beggars and Vagrants to see the trend**

```python
plt.hist(Total_census_data_State_wise["Main Activity Beggars, Vagrants etc. - Persons"], bins= 30, color ='skyblue')
plt.ticklabel_format(style='plain')
plt.tick_params(axis='x', labelsize=8)
plt.xlabel("Total number of Beggars and Vagrants")
plt.ylabel("Counts")
plt.title("Number of Beggars and Vagrants count in all the states in India till the year of 2011")
plt.show()
```

### **To find the 1st, 2nd,3rd and 4th quantile for the variable Beggars and Vagrants**

```python
Total_census_data_State_wise["Main Activity Beggars, Vagrants etc. - Persons"].quantile([0.1,0.5,0.75,1])
```

### **Plotting the Kernal density estimate plot for the variable**

```python
sns.kdeplot(data = Total_census_data_State_wise["Main Activity Beggars, Vagrants etc. - Persons"])
plt.ticklabel_format(style='plain')
plt.xlabel("Total number of Beggars and Vagrants")
plt.ylabel("Probability density of the people")
plt.title(" Kernal density estimate plot for the number of Beggars and Vagrants of all states ")
plt.show()
```

### **Plotting the box plot to observe the outliers for Beggars and Vagrants column**

```python
sns.boxplot(x = Total_census_data_State_wise["Main Activity Beggars, Vagrants etc. - Persons"])
plt.xlabel("Total number of Beggars and Vagrants")
plt.title(" Box plot for the number of Beggars and Vagrants of all states to find the outliers ")
plt.show()
```
