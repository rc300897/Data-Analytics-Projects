### **Importing the Packages**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

### **Reading the data from the csv file**

```python
census_data= pd.read_csv("census_2011_ddw_0000b_15_2011.csv")
```

```python
census_data
```

### **Creating a subset of data for total count of persons for all states and for all educational levels**

```python
Total_census_data = census_data[(census_data["Total/Rural/Urban"]=="Total")]
Total_census_data
```

### **Removing the first 7 rows since it has data for whole country and not states**

```python
Total_census_data = Total_census_data.iloc[8: , : ]
Total_census_data
```

### **Creating a index to remove the rows where educational level = Total**

```python
indexnames = Total_census_data[ Total_census_data['Educational Level'] == 'Total' ].index
indexnames
```

###  **Removing the rows from the dataframe where educational level is Total**

```python
Total_census_data.drop(indexnames, axis = 0, inplace=True)
Total_census_data
```

### **Finding the total no of rows and columns in the data**

```python
Total_census_data.shape
```

### **Observing the first five rows of the dataframe**

```python
Total_census_data.head()
```

### Observing the last 5 rows of the dataset**

```python
Total_census_data.tail()
```

### **To find the datatypes of all th columns and thier information**

```python
Total_census_data.info()
```

### **Summarizing the data to find various statistical variables to observe trends**

```python
Total_census_data.describe()
```

**We are going to look the column Non Workers persons for all the educational levels included and perform the EDA**

### **To find maximum value in the Total no of Non Working persons**

```python
Total_census_data["Non-Workers - Persons"].max()
```

### **To find minimum value in the Total no of Non Workers**

```python
Total_census_data["Non-Workers - Persons"].min()
```

### **Finding the mean of the Non working  persons**

```python
Total_census_data["Non-Workers - Persons"].mean()
```

### **Finding the median of the Non working  persons**

```python
Total_census_data["Non-Workers - Persons"].median()
```

### **Finding the Standard deviation of the Non working  persons**

```python
Total_census_data["Non-Workers - Persons"].std()
```

### **Plotting the histogram for total no of non working  persons to see the trend**

```python
plt.hist(Total_census_data["Non-Workers - Persons"], bins= 15, color ='skyblue',edgecolor='Black',linewidth=1)
plt.ticklabel_format(style='plain')
plt.tick_params(axis='x', labelsize=8)
plt.xlabel("Total No of non working Persons")
plt.ylabel("Counts")
plt.title("Histogram showing the distribution of  non working persons of all states")
plt.show()
```

### **To find the 1st, 2nd,3rd and 4th quantile for the variable Non working persons**

```python
Total_census_data["Non-Workers - Persons"].quantile([0.1,0.5,0.75,1])
```

### **Plotting the Kernal density estimate plot for the variable**

```python
sns.kdeplot(data = Total_census_data["Non-Workers - Persons"])
plt.ticklabel_format(style='plain')
plt.xlabel("Total No of non working  Persons")
plt.ylabel("Probability density of the persons")
plt.title(" Kernal density estimate plot for the non working  persons of all states ")
plt.show()
```

### **Plotting the box plot to observe the outliers for our Non working  persons column**

```python
sns.boxplot(x= Total_census_data["Non-Workers - Persons"])
plt.xlabel("Total No of non working  Persons")
plt.title(" Box plot for the No of non working  persons of all states to find the outliers ")
plt.show()
```
